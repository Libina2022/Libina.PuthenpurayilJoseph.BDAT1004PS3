**Step 1. Import the necessary libraries**

```python
import pandas as pd
```

**Step 2. Import the dataset from this address,  Step 3. Assign it to a variable called users**

```python
users = pd.read_csv('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user', sep='|')
```

```python
users.head()
```

**Step 4. Discover what is the mean age per occupation**

```python
users.groupby('occupation')[['age']].mean()
```

**Step 5. Discover the Male ratio per occupation and sort it from the most to the least**

```python
maleUsers = users[users['gender'] == 'M']
mCount=maleUsers.groupby(['occupation','gender'])['gender'].count()
totalCount=users.groupby(['occupation'])['gender'].count()
ratio=(mCount/totalCount) * 100
ratio.sort_values(ascending=False)
```

**Step 6. For each occupation, calculate the minimum and maximum ages**

```python
users.groupby('occupation').age.agg(['min', 'max'])
```

**Step 7. For each combination of occupation and sex, calculate the mean age**

```python
users.groupby(['occupation','gender'])['age'].mean()
```

**Step 8. For each occupation present the percentage of women and men**

```python
Count=users.groupby(['occupation','gender'])['gender'].count()
totalCount=users.groupby(['occupation'])['gender'].count()
ratio=((Count/totalCount)*100)
ratio
```
